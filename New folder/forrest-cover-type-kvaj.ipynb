```python
import pandas as pd


# Read the CSV file into a DataFrame named 'forrest'
train = pd.read_csv('/kaggle/input/forrestcovertype/train.csv')

# Display the first few rows of the DataFrame to verify the data has been loaded
print(train.head())
```

```python
train.info()
```

```python
#check if there is any class imbalancing
train['Cover_Type'].value_counts(normalize= True)
```

```python
from sklearn.model_selection import train_test_split

X_train = train.drop(columns=['Cover_Type'])  # Features (independent variables)
y_train = train['Cover_Type']  # Target variable (dependent variable)

# Print the shape of the resulting sets to verify the split
print("Training set shape:", X_train.shape, y_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
```

```python
from keras.utils import to_categorical

# Assuming y_train is a pandas Series
y_train_encoded = to_categorical(y_train - 1)  # Subtracting 1 to convert class labels from 1 to 7 to 0 to 6
```

**Build an Artificial Neaural Network**

```python
import tensorflow as tf

#initialize
ann = tf.keras.models.Sequential()
```

```python
#input layer
ann.add(tf.keras.layers.Dense(units=50, activation='relu'))
```

```python
#second layer
ann.add(tf.keras.layers.Dense(units=50, activation='relu'))
```

```python
#output layer
ann.add(tf.keras.layers.Dense(units=7, activation='softmax'))
```

**Training**

```python
ann.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = ann.fit(X_train, y_train_encoded, batch_size = 32, epochs = 100)
```

```python
import matplotlib.pyplot as plt

# Plot training & validation accuracy values
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.title('Model accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
```

```python
# Save the model to a file
ann.save("forrest_ann.h5")
```
